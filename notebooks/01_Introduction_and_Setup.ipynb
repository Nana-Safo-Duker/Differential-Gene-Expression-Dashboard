## ðŸ§¬ What is Differential Gene Expression?

**Differential gene expression analysis** identifies genes whose expression levels differ significantly between experimental conditions (e.g., treated vs. control, diseased vs. healthy).

### Key Concepts

- **log2 Fold Change**: Measures the magnitude of expression change
  - Positive values = upregulated in condition B vs. A
  - Negative values = downregulated in condition B vs. A
  - log2FC of 1 = 2-fold increase; log2FC of -1 = 2-fold decrease

- **P-value**: Statistical significance of the change
  - Raw p-value from statistical test
  - Adjusted p-value (padj/FDR): corrected for multiple testing

- **Common Thresholds**:
  - |log2FC| â‰¥ 1 (2-fold change)
  - Adjusted p-value < 0.05
